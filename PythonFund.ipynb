```python
#List are ordered, mutable and sequenced data types in Python
lt=[1,2,"vbas"]
print(lt[2])
#adding new elements to an existing list
lt.append("Pro")
print(lt)
#adding new elements to an existing list at a specific position
lt.insert(1,"Bab")
print(lt)
#removing an element from a list
lt.remove("Pro")
print(lt)
#removing an element from a list at a specific position
lt.pop(2)
print(lt)
#merging two list
lt1=[1,2,3]
lt2=[4,5,6]
lt1.extend(lt2)
print(lt1)
```

```python
#Tuples are immutable,ordered and sequenced data types in Python
tp=(1,2,"Pro")
print(tp[1])
```

```python
#sets are unordered,mutable storing unique elements. Hence duplicate elements will be removed
set1={1,2,3,4,4,"Pro","Pro","Hy"}
print(set1)
for i in set1:
  print(set1)
```
